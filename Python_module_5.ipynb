```python
## Question 10: Nested Lists
students = []

for _ in range(int(input())):
    name = input()
    score = float(input())
    students.append([name, score])

scores = sorted(set(score for name, score in students))
second_lowest = scores[1]

names = sorted(name for name, score in students if score == second_lowest)

for name in names:
    print(name)
```

```python
## Question 11: Finding the Percentage


n = int(input())
student_marks = {}

for _ in range(n):
    name, *line = input().split()
    scores = list(map(float, line))
    student_marks[name] = scores

query_name = input()

average = sum(student_marks[query_name]) / len(student_marks[query_name])
print("{:.2f}".format(average))
```

```python
## Question 12: Lists


N = int(input())
lst = []

for _ in range(N):
    command = input().split()

    if command[0] == "insert":
        lst.insert(int(command[1]), int(command[2]))
    elif command[0] == "print":
        print(lst)
    elif command[0] == "remove":
        lst.remove(int(command[1]))
    elif command[0] == "append":
        lst.append(int(command[1]))
    elif command[0] == "sort":
        lst.sort()
    elif command[0] == "pop":
        lst.pop()
    elif command[0] == "reverse":
        lst.reverse()
```

```python
## Question 18: Find a String

def count_substring(string, sub_string):
    count = 0

    for i in range(len(string) - len(sub_string) + 1):
        if string[i:i+len(sub_string)] == sub_string:
            count += 1

    return count

string = input()
sub_string = input()

print(count_substring(string, sub_string))


```

```python
## Question 20: Text Alignment
thickness = int(input())
c = 'H'

for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))

for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
## Question 23: String Formatting
def print_formatted(number):
    width = len(bin(number)) - 2

    for i in range(1, number + 1):
        print(
            str(i).rjust(width),
            oct(i)[2:].rjust(width),
            hex(i)[2:].upper().rjust(width),
            bin(i)[2:].rjust(width)
        )

n = int(input())
print_formatted(n)
```

```python
## Question 24: Alphabet Rangoli
def print_rangoli(size):
    import string

    alpha = string.ascii_lowercase
    width = 4 * size - 3

    for i in range(size - 1, -size, -1):
        x = abs(i)
        row = "-".join(alpha[size-1:x:-1] + alpha[x:size])
        print(row.center(width, "-"))

n = int(input())
print_rangoli(n)
```

```python
## Question 26: The Minion Game
def minion_game(string):
    vowels = "AEIOU"
    kevin = 0
    stuart = 0
    n = len(string)

    for i in range(n):
        if string[i] in vowels:
            kevin += n - i
        else:
            stuart += n - i

    if kevin > stuart:
        print("Kevin", kevin)
    elif stuart > kevin:
        print("Stuart", stuart)
    else:
        print("Draw")

s = input()
minion_game(s)
```

```python
## Question 27: Merge the Tools
def merge_the_tools(string, k):
    for i in range(0, len(string), k):
        part = string[i:i+k]

        unique = ""
        for char in part:
            if char not in unique:
                unique += char

        print(unique)

string = input()
k = int(input())

merge_the_tools(string, k)
```

```python
## Question 35: Day 8 - Dictionaries and Maps
n = int(input())

phone_book = {}

for _ in range(n):
    name, number = input().split()
    phone_book[name] = number

q = int(input())  # number of queries

for _ in range(q):
    query = input()

    if query in phone_book:
        print(query + "=" + phone_book[query])
    else:
        print("Not found")
```
